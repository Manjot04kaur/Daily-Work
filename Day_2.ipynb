**Tuple Method**

In Python, a tuple is an ordered, immutable collection of elements. Tuples are similar to lists, but the key difference lies in their immutability.
Once a tuple is created, its elements cannot be changed or modified. This makes tuples suitable for situations where data should remain
constant throughout the program.

```
tuple_name = (element)
```
